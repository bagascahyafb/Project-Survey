```python
import scipy.stats as stats
#Buat fungsi untuk menghitung korelasi dan menguji korelasi
def lwise_corr_pvalues(df, to_csv= False, file_name = None, method= None):
    df = df.dropna(how= 'any')._get_numeric_data()
    dfcols = pd.DataFrame(columns=df.columns)
    rvalues = dfcols.transpose().join(dfcols, how='outer')
    pvalues = dfcols.transpose().join(dfcols, how='outer')
    length = str(len(df))

    if method == None:
        test = stats.pearsonr
        test_name = "Pearson"
    elif method == "spearman":
        test = stats.spearmanr
        test_name = "Spearman Rank"
    elif method == "kendall":
        test = stats.kendalltau
        test_name = "Kendall's Tau-b"

    for r in df.columns:
        for c in df.columns:
            rvalues[r][c] = round(test(df[r], df[c])[0], 4)

    for r in df.columns:
        for c in df.columns:
            pvalues[r][c] = format(test(df[r], df[c])[1], '.4f')
            
    if to_csv == False:
        print("Correlation test conducted using list-wise deletion", 
        "\n", "Total observations used: ",length, "\n", "\n", 
        f"{test_name} Correlation Values", "\n", rvalues,
        "\n", "Significant Levels", "\n", pvalues)
    if to_csv == True:
        print("Correlation test conducted using list-wise deletion", 
        "\n", "Total observations used: ",length, "\n", "\n", 
        f"{test_name} Correlation Values", "\n", rvalues,
        "\n",
        "Significant Levels", "\n", pvalues)
    
    file = open(file_name, 'a')
    file.write("Correlation test conducted using list-wise deletion" + "\n" + f"{test_name} correlation values" + "\n")
    file.write("Total observations used: " + length + "\n")
    file.close()
    rvalues.to_csv(file_name, header= True, mode= 'a')
    file = open(file_name, 'a')
    file.write("p-values" + "\n")
    file.close()
    pvalues.to_csv(file_name, header= True, mode= 'a')

lwise_corr_pvalues(dfq)
```

# Factor Analyzer

```python
# Import required libraries
import pandas as pd
from sklearn.datasets import load_iris
from factor_analyzer import FactorAnalyzer
import matplotlib.pyplot as plt
```

```python
from factor_analyzer.factor_analyzer import calculate_bartlett_sphericity
chi_square_value,p_value=calculate_bartlett_sphericity(dfq)
chi_square_value, p_value
```

```python
from factor_analyzer.factor_analyzer import calculate_kmo
kmo_all,kmo_model=calculate_kmo(dfq)
kmo_model
```

```python
# Create factor analysis object and perform factor analysis
fa = FactorAnalyzer()
fa.fit(dfq, 25)
# Check Eigenvalues
ev, v = fa.get_eigenvalues()
ev
```

```python
# Create scree plot using matplotlib
plt.scatter(range(1,dfq.shape[1]+1),ev)
plt.plot(range(1,dfq.shape[1]+1),ev)
plt.title('Scree Plot')
plt.xlabel('Factors')
plt.ylabel('Eigenvalue')
plt.grid()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

selected_columns = df[['Ketersediaan Tempat Ibadah', 'Kebersihan Tempat Ibadah', 'Kenyamanan Tempat Ibadah', 'Program dan Kegiatan Keagamaan']]

colors = ['b', 'g', 'r', 'black']
plt.figure(figsize=(12, 8))

for i, column in enumerate(selected_columns.columns):
    counts, bins, patches = plt.hist(selected_columns[column], bins=[1, 2, 3, 4, 5, 6], alpha=0.6, label=column, color=colors[i])
    
    for count, bin_edge, patch in zip(counts, bins, patches):
        plt.text(bin_edge + 0.5, count, str(int(count)), ha='center', va='bottom')

plt.xlabel('Skor')
plt.ylabel('Jumlah Responden')
plt.title('Distribusi Skor pada faktor Kebersihan tempat ibadah dan kegiatan keagamaan')
plt.xticks([1, 2, 3, 4, 5])
plt.legend()
plt.show()
```
